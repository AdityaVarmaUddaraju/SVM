```python
from helper.utils import get_iris_data, plot_decision_regions
```

```python
X_train_std, X_test_std, y_train, y_test = get_iris_data()
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(C=1, kernel="linear", random_state=1)
svm.fit(X_train_std, y_train)
```

```python
import numpy as np
X_combined_stack = np.vstack((X_train_std, X_test_std))
y_combined_stack = np.hstack((y_train, y_test))
plot_decision_regions(X_combined_stack, y_combined_stack, svm, test_idx=range(105, 150))
```

```python
svm = SVC(C=1, kernel="rbf", random_state=1, gamma=100.0)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_stack, y_combined_stack, svm, test_idx=range(105, 150))
```
